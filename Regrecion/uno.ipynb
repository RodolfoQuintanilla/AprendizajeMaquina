```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#Importación de métricas
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(200725)
```

```python
y = np.random.choice(range(300,1000), 100)
y_predict = np.random.choice(range(300,1000), 100)
```

```python
#mean absolite error
mean_absolute_error(y, y_predict)
```

```python
#root mean squared error -> mean squared error
nse = mean_squared_error(y, y_predict)
rmse = np.sqrt(nse)

print(f"nse: {nse}")
print(f"rmse: {rmse}")

```

```python
# mean absolute percentage error function: 1/N(sum((|y-y_predict|)/y)*100))
def mean_absolute_percentage_error(y, y_predict):
    return np.mean(np.abs((y - y_predict) / y)) * 100
```

```python
#mean absolute percentage error
mean_absolute_percentage_error(y, y_predict)
```
